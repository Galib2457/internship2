```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Set plotting style for clean, professional reports
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 1. LOAD THE DATASET
print("--- Step 1: Loading Dataset ---")
# Using the Titanic dataset as a standard, rich dataset for EDA
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print(f"Dataset successfully loaded! Shape: {df.shape[0]} rows, {df.shape[1]} columns.\n")


# 2. DATA INSPECTION & CLEANING
print("--- Step 2: Data Inspection & Summary Statistics ---")
# Display basic info about data types and missing values
print("\n### Data Info:")
print(df.info())

# Summary statistics for numeric features
print("\n### Summary Statistics (Numeric):")
print(df.describe().T)

# Check for missing values
print("\n### Missing Values Count:")
missing_vals = df.isnull().sum()
print(missing_vals[missing_vals > 0])

# Handling Missing Values (Basic Imputation)
# Fill missing Age with the median age
df["Age"] = df["Age"].fillna(df["Age"].median())
# Fill missing Embarked with the most common port
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
# Drop 'Cabin' due to excessive missing data for this basic analysis
df.drop(columns=["Cabin"], inplace=True, errors="ignore")
print("\nMissing values handled successfully.\n")


# 3. STATISTICAL SUMMARIES & DISTRIBUTIONS (Key Feature 1)
print("--- Step 3: Generating Visualizations ---")
os.makedirs("eda_outputs", exist_ok=True)

# Visualization 1: Distribution of Age (Univariate Analysis)
plt.figure()
sns.histplot(df["Age"], kde=True, color="skyblue", bins=30)
plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig("eda_outputs/01_age_distribution.png")
plt.close()
print("- Saved: eda_outputs/01_age_distribution.png")

# Visualization 2: Survival Count by Gender (Bivariate Analysis)
plt.figure()
sns.countplot(data=df, x="Survived", hue="Sex", palette="Set2")
plt.title("Survival Count Grouped by Gender (0 = No, 1 = Yes)")
plt.xlabel("Survived")
plt.ylabel("Passenger Count")
plt.savefig("eda_outputs/02_survival_by_gender.png")
plt.close()
print("- Saved: eda_outputs/02_survival_by_gender.png")


# 4. CORRELATION & IDENTIFYING INFLUENCING FACTORS (Key Feature 2)
print("\n--- Step 4: Correlation & Pattern Analysis ---")

# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

print("\n### Correlation Matrix with Target (Survived):")
print(corr_matrix["Survived"].sort_values(ascending=False))

# Visualization 3: Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.savefig("eda_outputs/03_correlation_heatmap.png")
plt.close()
print("- Saved: eda_outputs/03_correlation_heatmap.png")


# 5. KEY INSIGHT GENERATION
print("\n--- Step 5: Key Insights Extracted ---")
# Calculate explicit insights to back up your report
survival_rate_gender = df.groupby("Sex")["Survived"].mean() * 100
survival_rate_class = df.groupby("Pclass")["Survived"].mean() * 100

print(
    f"1. Gender Influence: Female survival rate was {survival_rate_gender['female']:.2f}% "
    f"compared to male survival rate of {survival_rate_gender['male']:.2f}%."
)
print(
    f"2. Socio-Economic Factor: Class 1 passengers had a {survival_rate_class[1]:.2f}% survival rate, "
    f"while Class 3 passengers had only a {survival_rate_class[3]:.2f}% survival rate."
)
print("3. Correlation: Passenger Class (Pclass) shares a strong negative correlation with survival (-0.34).")
```
